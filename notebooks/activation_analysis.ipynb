# Time-course activation — Blind vs Sighted

Per-area activation (number of active neurons over time-steps) for two paradigms:

* **P1** — Phase 1, split into *Action* (patt_no 1-6) and *WF* / word-form
  (patt_no 7-12), analysed at presentation 5000.
* **P2** — Phase 2 (word patterns), analysed at presentation 50.

For each area the half-max response **window** is detected, activation is averaged
within it, and a Condition × System × Area repeated-measures ANOVA
(Greenhouse-Geisser corrected) plus per-area post-hoc t-tests are run.

If `data/activation_p1.csv` / `activation_p2.csv` are missing, build them once:

```bash
python scripts/build_activation.py --paradigm P1 --sighted-dir … --blind-dir …
python scripts/build_activation.py --paradigm P2 --sighted-dir … --blind-dir …
```

## Setup

````python
import sys
from pathlib import Path
sys.path.insert(0, str(Path.cwd().parent))

from activation import load_activation, activation_window, plotting
from activation.stats import activation_rm_anova, activation_posthoc

p1 = load_activation("P1")
p2 = load_activation("P2")
p1.head()
````

## 1. P1 time-course — Auditory Word Recognition (WF)

````python
win_wf = activation_window(p1, presentation=5000, spe="WF")
plotting.plot_timecourse_grid(p1, "P1", 5000, spe="WF", window=win_wf)
````

### P1 WF — statistics (RM-ANOVA + post-hoc)

````python
res_wf = activation_rm_anova(p1, win_wf, presentation=5000, spe="WF")
````

````python
posthoc_wf = activation_posthoc(res_wf["ab"])
````

## 2. P1 time-course — Action Execution

````python
win_act = activation_window(p1, presentation=5000, spe="Action")
plotting.plot_timecourse_grid(p1, "P1", 5000, spe="Action", window=win_act)
````

### P1 Action — statistics

````python
res_act = activation_rm_anova(p1, win_act, presentation=5000, spe="Action")
````

````python
posthoc_act = activation_posthoc(res_act["ab"])
````

## 3. P2 time-course (presentation 50)

````python
win_p2 = activation_window(p2, presentation=50)
plotting.plot_timecourse_grid(p2, "P2", 50, window=win_p2)
````

### P2 — statistics

````python
res_p2 = activation_rm_anova(p2, win_p2, presentation=50)
````

````python
posthoc_p2 = activation_posthoc(res_p2["ab"])
````
